```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40]
y = ['a','s','d','f']

plt.pie(x, labels= y, autopct= '%0.1f%%', explode= [0,0,0,0.1], radius= .8, textprops={"fontsize":10}, startangle= 90, labeldistance= 1.2, counterclock= False, wedgeprops={'linewidth' : .5}, rotatelabels= False)
plt.legend()
plt.show()
```

```python
x = [10,20,30,40]
y = ['a','s','d','f']
z = [0.01]

plt.pie(x, labels= y, autopct= '%0.1f%%', explode= [0,0,0,0.1], textprops={"fontsize":10}, startangle= 90, labeldistance= 1.2, counterclock= False, wedgeprops={'linewidth' : .5}, rotatelabels= False, radius= 1.5)
plt.pie(z, colors= 'w')
plt.legend()
plt.show()
```
