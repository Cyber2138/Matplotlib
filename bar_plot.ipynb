```python
import matplotlib.pyplot as plt
```

THIS IS THE VERTICAL BAR GRAPH 

```python
x = ['Python','C','C++','Java']
y = [100,50,90,85]
z = [80,10,33,45]

plt.bar(x,y, width = .5, alpha = .6, label = 'languages')
plt.bar(x,z, width = .5, alpha = .6, label = 'popularity')
plt.title('Data analysis of the Programming language popularity', fontsize = 12)
plt.xlabel('Programming languages', fontsize = 12)
plt.ylabel('Popularity', fontsize = 12)
plt.legend()
plt.show()
```

THIS IS THE HORIZONTAL BAR GRAPH

```python
x = ['Python','C','C++','Java']
y = [100,50,90,85]
z = [80,10,33,45]

plt.barh(x,y, alpha = .6, label = 'languages')
plt.barh(x,z, alpha = .6, label = 'popularity')
plt.title('Data analysis of the Programming language popularity', fontsize = 12)
plt.xlabel('Programming languages', fontsize = 12)
plt.ylabel('Popularity', fontsize = 12)
plt.legend()
plt.show()
```
