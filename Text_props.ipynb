```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [3,2,2,6,1]
plt.figure(figsize=(6,3))
plt.plot(x,y, color = 'black', marker = 'o')
plt.text(4.1,6, "Revenue", bbox = {'facecolor' : 'g', 'alpha': .5})
plt.annotate('java', xy=(2,2), xytext = (3,4), arrowprops= {'facecolor': 'red','shrink' : 100} )
plt.title('line graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend(['line'], loc = 9, facecolor = 'g', framealpha = .5)

plt.show()
```
