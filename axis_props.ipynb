```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [2,5,4,1,3]

# plt.figure(figsize=(5,3))
plt.plot(x,y, color = 'r', marker = 'o')
plt.title("This is the Line chart")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.xticks(x)
plt.yticks(y)
# plt.axis([0,10,0,10])
plt.xlim(0,7)
plt.show()
```
