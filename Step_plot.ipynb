```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,2,3,5,4]

plt.step(x,y, color = 'r', marker = 'o', markersize = 10, markerfacecolor = 'g', label = 'python')
plt.title('step plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend(loc = 2)
plt.grid()
plt.show()
```
