```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
plt.subplot(2,2,1)
plt.plot(x)
plt.show()

y = [6,5,4,3]
x = ['a','b','c','d']
plt.subplot(2,2,2)
plt.bar(x,y)
plt.show()

l = [20,30,40,50]
plt.subplot(2,2,3)
plt.pie(l)
plt.show()

plt.subplot(2,2,4)
plt.pie([1])
plt.show()
```
