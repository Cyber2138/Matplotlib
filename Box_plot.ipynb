```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40,50,60,70,120]
y = [100,80,37,50,20,10,66,88]
z = [x,y]
for i in z:
    plt.boxplot(i, patch_artist= True, showmeans= True, boxprops= {'color' :'r'}, whiskerprops= {'color': 'g'}, vert= True, labels= ['python'], sym= 'r+', capprops= {"color" : 'purple'}, flierprops= {'markeredgecolor': 'black'})

    plt.show()
```
